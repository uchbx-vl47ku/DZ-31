```python
import os
import shutil

ls = []

def Get_dir():
    global ls
    ls = []
    #ls = os.listdir(path='.')
    it = os.scandir(os.getcwd())
    for entry in it:
        ty='f'
        if entry.is_dir()    : ty='d'
        if entry.is_symlink(): ty='l'
        ls.append({"name":entry.name , "type":ty})
    def FunSortFiles(el) :
        if el["type"]=='d': return '1'+el["name"]
        if el["type"]=='l': return '2'+el["name"]
        return '3'+el["name"]
    ls.sort(key=FunSortFiles)       

def Show_ls_files():
    print('------------------------------>>>',os.getcwd(),'<<<')
    for i in range(0,len(ls)):
        print( str(i).rjust(3).replace(' ','0'), ls[i]["type"],ls[i]["name"] )
    print('------------------------------')

                
def List_dir():
    Show_ls_files() 

def Choice_file():
    pass

def Change_dir():
    di = input('Введите имя для chdir ')
    try: 
        os.chdir(di)
    except OSError as error:  
        print('Ошибка',error) 
    Get_dir()
    Show_ls_files() 
    
def Create_dir():
    di = input('Введите имя для mkdir ')
    try: 
        os.mkdir(di)
    except OSError as error:  
        print('Ошибка',error) 
    Get_dir()
    Show_ls_files() 
    
def Copy_file():
    src = input('Введите имя копируемого файла ')
    dest = input('Введите имя целевой папки ')
    try: 
        shutil.copyfile(src, dest)
    except OSError as error:  
        print('Ошибка',error) 
    Get_dir()
    Show_ls_files()
    
def Move_file():
    src = input('Введите имя перемещаемого файла ')
    dest = input('Введите имя целевой папки ')
    try: 
        os.rename(src,dest)
    except OSError as error:  
        print('Ошибка',error) 
    Get_dir()
    Show_ls_files()
    
def Del_file():
    di = input('Введите имя удаляемого файла ')
    try: 
        os.remove(di)
    except OSError as error:  
        print('Ошибка',error) 
    Get_dir()
    Show_ls_files() 

Get_dir()

print('= = = Работа с файлами = = =')
print('------------------------------')
print('1 - содержимое папки ')
#print('2 - выбрать папку/файл ')
print('3 - перейти в папку ')
print('4 - создать папку ')
print('5 - копировать ')
print('6 - переместить ')
print('7 - удалить ')
print('e - завершение ')
print('------------------------------')
# выбор операции
while True:
    while True:
        op = input('Выберите операцию: ')
        if op in ('1','2','3','4','5','6','7','e'):
            break
        print('Операция', '"'+op+'"' ,'не обрабатывается')
    if   op=='1': List_dir()
    elif op=='2': Choice_file()
    elif op=='3': Change_dir()
    elif op=='4': Create_dir()
    elif op=='5': Copy_file()
    elif op=='6': Move_file()
    elif op=='7': Del_file()
    elif op=='e':
        break
```
